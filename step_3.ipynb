```python
# Загружаем модули

import pickle
import lightgbm as lgb
import pandas as pd
```

```python
# Загружаем модель

f=open("./lgb_0.983326.pickle","rb")
lgb_model = pickle.load(f)
f.close()

lgb_model
```

```python
# Загружаем данные

data = pd.read_pickle('./data_0.983326.zip', compression='zip')
data.head(10)
```

```python
drop_all_columns=['.geo', 'area', 'crop', 'id', 'geolist', 'location', 'adress', 'adress_2',
             'geo_2', 'geo_3', 'geo_5', 'longitude', 'latitude', 'points']
```

```python
# Создаем массив данных на которых модель обучалась из первых 10 строк данных

X_data = data[data['crop']>-1].drop(drop_all_columns, axis = 1)[:10].to_numpy()
X_data
```

```python
# Получаем предсказания - 10 из 10, все верные

lgb_model.predict(X_data)
```
